```python


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.species
```

```python


sns.pairplot(iris)
```

```python

sns.pairplot(iris,hue='species')
```

```python


sns.barplot(data=iris,x='species',y='sepal_length')
plt.xlabel('species')
plt.ylabel('sepal_length')
plt.title("Avarage sepal length by species")
plt.show()
```

```python


sns.barplot(data=iris,x='species',y='sepal_length')
plt.xlabel('species')
plt.ylabel('sepal_length')
plt.title("Avarage sepal length by species")
plt.show()
```

```python

sns.barplot(data=iris,x='species',y='petal_length')
plt.xlabel('species')
plt.ylabel('petal_length')
plt.title("Avarage petal length by species")
plt.show()
```

```python


sns.barplot(data=iris,x='species',y='petal_width')
plt.xlabel('species')
plt.ylabel('petal_width')
plt.title("Avarage petal width by species")
plt.show()
```

```python



species = iris["species"]
sepal_length = iris["sepal_length"]
petal_length = iris["petal_length"]
sepal_width = iris["sepal_width"]
petal_width = iris["petal_width"]


plt.bar(species, sepal_length)
plt.bar(species, petal_length)

plt.xlabel("Species")
plt.ylabel("Length")
plt.title("Sepal vs Petal Length ")

plt.show()
```

```python



plt.barh(species, sepal_width)
plt.barh(species, petal_width)

plt.xlabel("Species")
plt.ylabel("Width")
plt.title("Sepal vs Petal Length ")

plt.show()
```

```python



plt.scatter(iris["sepal_length"], iris["petal_length"])

plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Iris Scatter Plot ")
plt.show()
```

```python


plt.scatter(iris["sepal_width"], iris["petal_width"])

plt.xlabel("Sepal width")
plt.ylabel("Petal width")
plt.title("Iris Scatter Plot ")
plt.show()
```

```python


plt.hist(iris["sepal_length"], bins=10, edgecolor="white")

plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Iris Sepal Length Histogram")

plt.show()
```

```python

fig,ax=plt.subplots(1,3,figsize=(12,3))
ax[0].scatter(iris["sepal_width"], iris["petal_width"])
ax[1].hist(iris["sepal_length"], bins=10, edgecolor="white")
ax[2].barh(species, petal_width)
plt.show()
```
