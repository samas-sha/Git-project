Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
#num=[7,5,6,3,9]
#print("maximum of the list :",min(num))
#print("minimum of the list :",max(num))
def list_range(list_number):
  minimum_value = list_number[0] #setting the first value of the list as minimum value
  maximum_value = list_number[0] #settung the first value of the list as maximum value

  for number in list_number:
    #looping through the entire list to check for maximum_value and minimum_value
    if(number >= maximum_value):
      maximum_value = number

    if(number <= minimum_value):
      minimum_value = number

  return(minimum_value,maximum_value)

my_list =[66,88,77,54,3,4,99,74]
range = list_range(my_list)
range
```

 Q3. Write a function that returns the median value of a list of numbers

```python
num=[7,5,6,3,9]
def list_median(list_num):
  num = sorted(list_num)
  center = len(list_num) // 2
  if len(list_num) % 2 ==1:
    return num[center]

median = list_median(num)
median
```
